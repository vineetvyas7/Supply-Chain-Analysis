### Relationship Between Price and Revenue Generated

```python
fig = px.scatter(df,x='Price',y='Revenue generated',color='Product type',hover_data=['Number of products sold'],trendline='ols')
fig.show()
```

### Sales Distribution by Product Type

```python
sales_data = df.groupby('Product type')['Number of products sold'].sum().reset_index()

pie_chart  = px.pie(sales_data, values = 'Number of products sold',
                    names = 'Product type',title='Sales by Product Type',
                    hover_data=['Number of products sold'],hole=0.5,
                    color_discrete_sequence=px.colors.qualitative.Pastel)

pie_chart.update_traces(textposition='inside',textinfo='percent+label')
pie_chart.show()
```

### Defect Rate Distribution by Product Type

```python
defect_rate = df.groupby('Product type')['Defect rates'].sum().reset_index()

pie_chart = px.pie(defect_rate,values='Defect rates',
                   names= 'Product type',
                   title='Defect Data Rates',
                   hover_data= ['Defect rates'],hole=0.5,
                   color_discrete_sequence=px.colors.qualitative.Pastel)
pie_chart.update_traces(textposition='inside',textinfo='percent+label')
pie_chart.show()
```

### Transportation Cost Distribution by Mode

```python
transportation_cost = df.groupby('Transportation modes')['Costs'].sum().reset_index()
pie_chart = px.pie(transportation_cost,values='Costs',
                   names='Transportation modes',
                   title='Cost Distribution by Transportantion Mode',
                   hover_data=['Costs'],hole=0.5,
                   color_discrete_sequence=px.colors.qualitative.Pastel)
pie_chart.update_traces(textposition='inside',textinfo='percent+label')
pie_chart.show()
```

### Total Revenue by Shipping Carrier

```python
total_revenue = df.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
fig = go.Figure()
fig.add_trace(go.Bar(x=total_revenue['Shipping carriers'],
                     y=total_revenue['Revenue generated']))

fig.update_layout(title='Total Revenue by Shipping Carrier',
                  xaxis_title = 'Shipping Carrier',
                  yaxis_title = 'Revenue Generated')
fig.show()
```

### Average Lead Time and Manufacturing Cost by Product Type

```python
avg_lead_time=df.groupby('Product type')['Lead time'].mean().reset_index()
#avg_lead time
avg_manufacturing_costs = df.groupby('Product type')['Manufacturing costs'].mean().reset_index()
result = pd.merge(avg_lead_time,avg_manufacturing_costs, on ='Product type')
result.rename(columns={'Lead time':'Average Lead Time',
                       'Manufacturing costs':'Average Manufacturing costs'},
              inplace=True)
print(result)
```

```python
fig = px.bar(
    result,
    x='Product type',
    y=['Average Lead Time', 'Average Manufacturing costs'],
    barmode='group',
    title='Average Lead Time and Manufacturing Cost by Product Type'
)

fig.show()
```

## Revenue Analysis by Product Type

```python
import plotly.express as px

# Revenue by Product Type
revenue_by_product = (
    df.groupby("Product type", as_index=False)["Revenue generated"]
      .sum()
      .sort_values("Revenue generated", ascending=False)
)

fig = px.bar(
    revenue_by_product,
    x="Product type",
    y="Revenue generated",
    color="Revenue generated",
    color_continuous_scale="Blues",
    text_auto=".2s",
    title="Revenue Generated by Product Type",
    labels={
        "Product type": "Product Type",
        "Revenue generated": "Total Revenue"
    },
    template="plotly_white"
)

fig.update_layout(
    title_x=0.5,
    xaxis_title="Product Type",
    yaxis_title="Total Revenue",
    margin=dict(t=80)
)

fig.update_yaxes(
    range=[0, revenue_by_product["Revenue generated"].max() * 1.15]
)

fig.update_traces(textposition="outside")
fig.show()
```
